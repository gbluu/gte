```python
import pandas as pd
import os
import glob
import numpy as np

# setting notebook
pd.set_option('display.max_columns', None)
```

```python
dmsp = pd.read_csv(r'../../data/raw/03-dmsp/dmsp.csv', index_col=None, header=4)
```

```python
dmsp.head(1)
```

```python
# Rename columns
dmsp_rename = dmsp.rename(
    columns={
        'BRAND': 'brand',
        'PRODUCT CODE': 'product_code',
        'PRODUCT NAME': 'product_name',
        'CBM/Unit': 'cbm_unit',
        'Mã chủ hàng': 'seller'
    }
)
```

```python
#Drop null rows
dmsp_drop_null = dmsp_rename.dropna(subset=['product_code']).reset_index(drop=True) 
```

```python
#convert data type
dmsp_convert = dmsp_drop_null.copy()
dmsp_convert['GrossWeightProductg'] = dmsp_convert['GrossWeightProductg'].str.strip().str.replace(',','').replace("-", "0").astype(float)
dmsp_convert['cbm_unit'] = dmsp_convert['cbm_unit'].str.strip().str.replace("-", "0").astype(float)
```
